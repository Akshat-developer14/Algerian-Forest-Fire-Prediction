# EDA - Exploratory Data Analysis of our new cleaned dataset of Algerian forest fires

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('../data/Algerian_forest_fires_cleaned_dataset.csv')
```

```python
dataset.head()
```

```python
ds_copy = dataset.drop(['day', 'month', 'year'], axis=1)
```

```python
ds_copy.head()
```

- Sometimes in data categories can be different due to space even if actually they are same

```python
ds_copy['Classes'].value_counts()
```

```python
# Encoding of the categories in classes
ds_copy['Classes'] = np.where(ds_copy['Classes'].str.contains('not fire'), 0, 1)
```

```python
ds_copy.head()
```

```python
ds_copy.tail()
```

```python
ds_copy['Classes'].value_counts()
```

```python
## Plot Density plot for all features
ds_copy.hist(bins=50, figsize=(20, 15))
```

```python
# Percentage for pie chart
percentage = ds_copy['Classes'].value_counts(normalize=True)*100
percentage
```

```python
# plotting pie chart
classlabels = ['Fire', 'Not Fire']
plt.figure(figsize=(12, 7))
plt.pie(percentage, labels=classlabels, autopct='%1.1f%%')
plt.title('Pie chart of classes')
```

```python
# correlation

ds_copy.corr()
```

```python
sns.heatmap(ds_copy.corr())
```

```python
# Box Plots
sns.boxplot(x = 'FWI', data=dataset, color='g')
```

```python
dataset.head()
```

```python
# Encoding of the categories in classes
dataset['Classes'] = np.where(dataset['Classes'].str.contains('not fire'), 'not fire', 'fire')
```

```python
# Monthly fire analysis
ds_temp = dataset.loc[dataset['Region']==1]
plt.figure(figsize=(13, 6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=dataset)
plt.xlabel('Number of fires', weight='bold')
plt.ylabel('Months', weight='bold')
plt.title('Fire Analysis of Sidi-Bel Regions', weight='bold')
```

```python
# Monthly fire analysis
ds_temp = dataset.loc[dataset['Region']==0]
plt.figure(figsize=(13, 6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=dataset)
plt.xlabel('Number of fires', weight='bold')
plt.ylabel('Months', weight='bold')
plt.title('Fire Analysis of Bejaia Regions', weight='bold')
```

Its observed that August and September had the most number of forest fires for both regions. And from the above plot of months, we can understand few things

Most of the fires happened in August and very high Fires happened in only 3 months - June, July and August.

Less Fires was on September
